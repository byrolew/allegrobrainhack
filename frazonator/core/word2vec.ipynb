```python
import numpy as np
from gensim.models import KeyedVectors
```

```python
model = KeyedVectors.load_word2vec_format('vectors.txt', binary=False, unicode_errors='replace')
```

```python
for word, score in model.most_similar(positive=['torebka'], negative=[], topn=1000):
    print(score, word)
```
